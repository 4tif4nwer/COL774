```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import tree
```

```python
full_data = pd.read_csv("D:/GitHub/COL774/Assignment 3/data/col774_mammography/train.csv")
```

```python
filtered_data = full_data[(full_data["Age"]!='?') & (full_data["Shape"]!='?') & (full_data["Margin"]!='?') & (full_data["Density"]!='?')]
```

```python
x_train_data = filtered_data[["Age","Shape","Margin","Density"]].copy()
y_train_data = filtered_data[["Severity"]].copy()

print(x_train_data)
print(y_train_data)
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(x_train_data, y_train_data)
```

```python
tree.plot_tree(clf, feature_names=["Age","Shape","Margin","Density"],)
```
